```python
def OP(power):

    if power < 320:
        return "*"
    elif power < 421:
        return "**"
    elif power < 500:
        return "***"
    else:
        return "****"
```

```python
df["Star"] = df["Total"].apply(OP)
```

```python
df.iloc[456:465]
```

```python
def defense_power(Def, Sp_Def):
    if Def + Sp_Def > 230:
        return "Shield"
    else:
        return "Neutral"
```

```python
df["Defender"] = df[["Defense", "Sp. Def"]].apply(lambda df: defense_power(df["Defense"], df["Sp. Def"]), axis=1)
```

```python
df[df["Defender"] == "Shield"]
```

```python
def attack_power(Atk, Sp_Atk):
    if Atk + Sp_Atk > 240:
        return "Sword"
    else:
        return "Neutral"
```

```python
df["Attacker"] = np.vectorize(attack_power)(df["Attack"],df["Sp. Atk"])
```

```python
df[df["Attacker"] == "Sword"]
```

### .apply(lambda) VS np.cectorize()

```python
import timeit 

setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('pokemon.csv')
def defense_power(Def, Sp_Def):
    if Def + Sp_Def > 230:
        return "Shield"
    else:
        return "Neutral"
'''
  
stmt_one = ''' 
df["Defender"] = df[["Defense", "Sp. Def"]].apply(lambda df: defense_power(df["Defense"], df["Sp. Def"]), axis=1)
'''

stmt_two = '''
df["Defender"] = np.vectorize(defense_power)(df["Attack"], df["Sp. Atk"])
'''
# Bu kodu 2 defa √ßalƒ±≈ütƒ±rƒ±rsan patlar. bir defa √ßalƒ±≈ütƒ±r ve a≈üaƒüƒ± ge√ß
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```

Kƒ±saca **np.vectorize()** tam bir yok edici
