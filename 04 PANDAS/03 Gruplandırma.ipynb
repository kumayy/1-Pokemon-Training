# **Gruplandırma Operasyonları**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("cars.csv")
```

## .groupby()

* .mean(): Grubun ortalaması
* .sum(): Grubun toplamı
* .size(): Grup boyutu
* .count(): Grup sayısı
* .min(): min grup değeri
* .max(): max grup değeri
* .describe(): Aynısı
* .std(): Standart sapma
* .var(): Varyans
* .sem(): Standard error of the mean
* .first(): Gruptaki ilk değer
* .last(): Gruptaki son değer

```python
df.head()
```

Başlıklar:

        mpg: Miles per gallon
        cyl: Silindir
        disp: Deplasman (Cubic inch)
        hp: Horse power
        drat: Rear axle ratio
        wt: Weight(1000 lb)
        qsec: 1/4 mil süresi
        vs: Motor (V shaped = 0, straight = 1)
        am: Transmisyon (0 = Auto, 1 = Man)
        gear: Vites
        carb: Karbüratör sayısı

```python
df.groupby("cyl")
```

```python
df.groupby("cyl").mean()
```

## Aggregate

```python
# help(df.agg)
```

```python
df.agg(["mean","max","min"])
```

```python
df.agg(["mean","max","min"])[["mpg","hp","gear"]]
```

```python
df.agg({
    "hp" : ["max","min"],
    "mpg" : ["mean","median","std"],
    "wt" : ["sum"]
})
```

```python
df.groupby("cyl").agg({
    "hp" : ["max","min"],
    "wt" : ["sum"]
})
```
