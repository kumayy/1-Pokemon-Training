## Çok Efsane Bir Yöntem Daha


###### Eksik veriyi başka gruptaki verilerden bağlantı kurarak doldurma

```python
df
```

```python
df["mass"] = df["mass"].fillna(df.groupby("sex")["mass"].transform("mean"))
```

```python
df
```

```python
df["height"] = df["height"].fillna(df.groupby(["sex","species"])["height"].transform("mean"))
```

```python
df
```

```python
df.dropna(subset=["sex","species"])
```
