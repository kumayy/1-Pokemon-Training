### Pivot Tablo Ne İçin Kullanılır?

* Çok miktarda veriyi, kullanımı kolay birden çok şekilde sorgulama,
* Sayısal verilerin alt toplamını ve toplamını alma, kategori ve alt kategorilerle verileri özetleme ve özel hesaplamalarla formül oluşturma,
* Sonuçlara odaklanmak için veri düzeylerini genişletme ve daraltma, ilgi alanlarına göre özet verilerden ayrıntılara ulaşma,
* Kaynak verilerin farklı özetlerini görmek amacıyla satırları sütunlara veya sütunları satırlara taşıma veya özetleme,
* İstenen bilgilere odaklanmaya olanak sağlamak amacıyla verilerin en yararlı ve ilginç alt kümesini sıralama, filtre uygulama, gruplandırma ve koşullu biçimlendirme,
* Doğru, ilginç ve ek açıklamalı çevrimiçi veya basılı raporları sunma.

## .pivot_table()

Çok detaylı bilgi için: https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html

Soru: **Pokemon türlerinin her nesile göre ortalama saldırı değerleri nedir?**

Soru: **Türlerin nesillere göre ortalama saldırı güçleri ne kadar değişiyor?**

Sana Özel Soru: **1. nesil böcek pokemonlarının ortalama saldırısı nedir?**
###### Cevap aşağıda ama kaydırmadan önce bi sen dene ^.^

```python
# help(pd.pivot_table)
```

```python
pd.pivot_table(data=df,index="Type 1",columns="Generation",values="Attack")
# .pivot_table( belirtilmediği sürece ortalama değeri verir )
```

```python
df[(df["Type 1"] == "Bug") & (df["Generation"] == 1)].mean()[3]
```

```python
pd.pivot_table(data=df,index="Generation",columns="Type 1",values="Attack")
```

```python
pd.pivot_table(data=df,index=["Generation","Legendary"],columns="Type 1",values="Attack")
```

```python
pd.pivot_table(data= df, index="Generation", aggfunc="sum")
```

```python
pd.pivot_table(df, index="Type 1", aggfunc="mean")
```

```python
pd.pivot_table(df, index="Type 1", aggfunc="mean")["Attack"]
```

```python
pd.pivot_table(df, index="Type 1", aggfunc="mean")[["Attack","Defense","Sp. Atk","Sp. Def"]]
```

```python
pd.pivot_table(df, index="Type 1",values=["Generation","Speed"], aggfunc="mean")
```

Tamam da bunun aynısını biz groupby metoduyla da yapıyorduk. Neden pivot tablo kullanalım ki?

## Pivot Tabloda Eksik veriler

```python
pd.pivot_table(data=df,index=["Generation","Legendary"],columns=["Type 1"],values=["Attack"], aggfunc="mean")
```

```python
pd.pivot_table(data=df,index=["Generation","Legendary"],columns=["Type 1"],values=["Attack"], aggfunc="mean", fill_value= 0)
```

```python
round(pd.pivot_table(data=df,index=["Type 1"],columns=["Generation"],values=["Attack"], aggfunc=["mean","max","min"], fill_value= 0),2)
```
