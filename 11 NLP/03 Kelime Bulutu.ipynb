```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
text = open('hobbit.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
wc = WordCloud().generate(text)
```

```python
plt.imshow(wc)
plt.axis("off")
```

---
## Alengirli Versiyon

```python
custom_mask = np.array(Image.open('dragon.png')) 
```

```python
wc = WordCloud(
    background_color= 'white',
    stopwords= stopwords,
    max_words= 800,
    mask= custom_mask,
    contour_width= 1,
    contour_color= 'black'
).generate(text)
```

```python
plt.figure(figsize=(12,12), dpi=200)
plt.imshow(wc)
plt.axis("off");
```

```python
wc.to_file("smaug.png")
```
