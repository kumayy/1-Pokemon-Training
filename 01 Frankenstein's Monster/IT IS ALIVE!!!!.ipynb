```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_excel("def_gdp.xlsx")
```

```python
df.head()
```

```python
X = df[["usd_exp_change"]]
y = df["usd_gdp_change"]
```

```python
# X = df["usd_exp_change"]
# y = df["usd_gdp_change"]
```

```python
X.shape
```

```python
regr = LinearRegression()
```

```python
type(X)
```

```python
regr.fit(X,y)
```

```python
df["usd_exp_change"].corr(df["usd_gdp_change"])
```

```python
print("Theta :     ", regr.coef_[0])
print("Intercept : ", regr.intercept_)
print("R^2 :       ",r2_score(X,y))
print("RMSE :      ",np.sqrt(mean_squared_error(X,y)))
print("MAE :       ",mean_absolute_error(X,y))
```

```python
X = X.to_numpy()
```

```python
X
```

```python
plt.figure(figsize=(10,6), dpi=200)
plt.plot(X, regr.predict(X), color="Red", lw=3)
sns.scatterplot(data=df, x="usd_exp_change", y="usd_gdp_change")

plt.xlabel("Savunma Harcaması Değişimi", size=12)
plt.ylabel("GSYİH Değişimi", size=12)
plt.title("Savunma Harcamalarının ve GSYİH'nin Yıllık Değişimi", size=15)
plt.grid(True)
plt.axvline(0, c='grey', ls="-")
plt.axhline(0, c='grey', ls="-")
plt.xlim(-0.4,0.6)
plt.ylim(-0.5,0.5);
```
